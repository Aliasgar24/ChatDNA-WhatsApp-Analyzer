# 🧬 ChatDNA
### Decode Your WhatsApp Conversations

**ChatDNA** is a WhatsApp Chat Analyzer built using Python.

It analyzes exported WhatsApp chats and generates detailed insights about:

- 📊 Chat Statistics
- 📅 Activity Timeline
- ⏰ Most Active Hours
- 💬 Word Analysis
- 😂 Emoji Analysis
- 🌶️ Slang & Gaali Analysis
- 🧠 Personality Detection
- ❤️ Relationship Insights
- 📈 Final Dashboard

**Dataset:** User's exported WhatsApp chat (.txt)


## 📚 Import Libraries

```python
import string
import numpy as np

from datetime import datetime
from collections import Counter
```

## 📖 Read and Parse WhatsApp Chat

```python
messages = []

with open("Whatsapp_GroupChat.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()
```

## 🔍 Parse WhatsApp Messages

```python
current_message = None

for line in lines:

    line = line.strip()

    if line == "":
        continue

    if len(line) >= 8 and line[2] == "/" and line[5] == "/":

        try:

            date_time, remaining = line.split(" - ", 1)

            sender, message = remaining.split(": ", 1)

            current_message = {
                "timestamp": date_time,
                "sender": sender,
                "message": message
            }

            messages.append(current_message)

        except:
            continue

    else:

        if current_message is not None:
            current_message["message"] += "\n" + line
```

```python
name_changes = {
    "Husain Vohra Alien 👾": "Husain Vohra",
    'Hussain Vohra' : 'Husain Vohra',
    'Mohammed Kalol New India' : 'Mohammed Kalol',
    'Mohammed Kalol Dubai' : 'Mohammed Kalol'
}
for message in messages:

    sender = message["sender"]

    if sender in name_changes:
        message["sender"] = name_changes[sender]
```

## 🧹 Clean and Classify Messages

```python
normal_messages = []

media_messages = []

deleted_messages = []

edited_messages = []

view_once_messages = []

poll_messages = []

system_messages = []

for message in messages:

    text = message["message"]

    if "<Media omitted>" in text:
        media_messages.append(message)

    elif "This message was deleted" in text:
        deleted_messages.append(message)

    elif "<This message was edited>" in text:
        edited_messages.append(message)
        normal_messages.append(message)

    elif "<View once" in text:
        view_once_messages.append(message)

    elif text.startswith("POLL:"):
        poll_messages.append(message)

    else:
        normal_messages.append(message)
```

## 📊 Chat Overview

```python
participants = set()

for message in messages:
    participants.add(message["sender"])

first_date = messages[0]["timestamp"].split(",")[0]
last_date = messages[-1]["timestamp"].split(",")[0]

days = (
    datetime.strptime(last_date, "%d/%m/%y") -
    datetime.strptime(first_date, "%d/%m/%y")
).days + 1

average_messages = len(messages) / days


print("\n" + "=" * 80)
print(" " * 28 + "🧬 CHAT OVERVIEW")
print("=" * 80)

print(f"\n💬 Total Messages      : {len(messages)}")
print(f"👥 Participants        : {len(participants)}")
print(f"📅 Chat Started        : {first_date}")
print(f"📅 Latest Message      : {last_date}")
print(f"📆 Chat Duration       : {days} days")

print()

print(f"🖼️ Media Messages      : {len(media_messages)}")
print(f"🗑️ Deleted Messages    : {len(deleted_messages)}")
print(f"✏️ Edited Messages     : {len(edited_messages)}")
print(f"🎙️ View Once Messages  : {len(view_once_messages)}")
print(f"📊 Polls               : {len(poll_messages)}")

print()

print(f"📈 Average Messages/Day: {average_messages:.2f}")

print("\n" + "=" * 80)
```

## 👥 Participant Analysis

```python
message_count = {}

for message in normal_messages:

    sender = message["sender"]

    if sender in message_count:
        message_count[sender] += 1
    else:
        message_count[sender] = 1

most_active = max(message_count, key=message_count.get)
least_active = min(message_count, key=message_count.get)




print("\n" + "=" * 90)
print(" " * 31 + "👥 PARTICIPANT ANALYSIS")
print("=" * 90)

print()

print(f"{'Participant':<25}{'Messages':<12}{'Percentage':<12}Contribution")
print("-" * 90)

for sender in sorted(message_count, key=message_count.get, reverse=True):

    percentage = (message_count[sender] / len(normal_messages)) * 100

    bar = "█" * int((percentage / 100) * 30)

    print(f"{sender:<25}{message_count[sender]:<12}{percentage:>6.2f}%     {bar}")
    print()

print("-" * 90)

print(f"🏆 Most Active  : {most_active} ({message_count[most_active]} messages)")
print(f"😴 Least Active : {least_active} ({message_count[least_active]} messages)")

print("\n" + "=" * 90)
```

## 📅 Timeline Analysis

```python
daily_messages = {}

monthly_messages = {}

yearly_messages = {}

for message in normal_messages:

    date = message["timestamp"].split(",")[0]

    if date in daily_messages:
        daily_messages[date] += 1
    else:
        daily_messages[date] = 1

    day, month, year = date.split("/")

    month_year = month + "/" + year

    if month_year in monthly_messages:
        monthly_messages[month_year] += 1
    else:
        monthly_messages[month_year] = 1

    if year in yearly_messages:
        yearly_messages[year] += 1
    else:
        yearly_messages[year] = 1



print("\n" + "=" * 90)
print(" " * 32 + "📅 TIMELINE ANALYSIS")
print("=" * 90)

print()

print(f"📆 Active Days    : {len(daily_messages)}")
print(f"📅 Active Months  : {len(monthly_messages)}")
print(f"🗓️ Active Years   : {len(yearly_messages)}")

print()

most_active_day = max(daily_messages, key=daily_messages.get)
least_active_day = min(daily_messages, key=daily_messages.get)

print(f"🔥 Most Active Day   : {most_active_day} ({daily_messages[most_active_day]} messages)")
print(f"😴 Least Active Day  : {least_active_day} ({daily_messages[least_active_day]} messages)")

print("\n" + "=" * 90)
```

## ⏰ Hourly Activity Analysis

```python
hourly_messages = {}

for hour in range(24):
    hourly_messages[hour] = 0

for message in normal_messages:

    time = message["timestamp"].split(",")[1].strip()

    dt = datetime.strptime(time, "%I:%M %p")

    hour = dt.hour

    hourly_messages[hour] += 1




print("\n" + "=" * 90)
print(" " * 29 + "⏰ HOURLY ACTIVITY")
print("=" * 90)

print()

max_messages = max(hourly_messages.values())

for hour in range(24):

    count = hourly_messages[hour]

    bar = "█" * int((count / max_messages) * 30)

    print(f"{hour:02d}:00 - {hour:02d}:59   {count:<6} {bar}")

print()

most_active_hour = max(hourly_messages, key=hourly_messages.get)

print(f"🔥 Peak Hour : {most_active_hour:02d}:00 - {most_active_hour:02d}:59")
print(f"💬 Messages  : {hourly_messages[most_active_hour]}")

print("\n" + "=" * 90)
```

## 🌶️ Slang & Gaali Analysis

```python
gaali_dict = {
    "BC": [
        "bc", "b.c", "b c",
        "behenchod", "behnchod", "bhenchod", "bhnchod"
    ],

    "MC": [
        "mc", "m.c", "m c",
        "madarchod", "maderchod", "madrchod", "madarchodh"
    ],

    "BH*SR*NA": [
        "bhosdi", "bhosdike", "bhosdiwala",
        "bhosrina", "bhosri"
    ],

    "LAVRA": [
        "lavra", "lavda", "lavdo"
    ],

    "L*ND": [
        "lund", "loda", "lodu", "lode"
    ],

    "CH*T*YA": [
        "chutiya", "chutiya", "chutiyo"
    ],

    "GANDU": [
        "gandu", "ganduu", "gando"
    ],

    "HARAMI": [
        "harami"
    ],

    "BAKCH*D": [
        "bakchod", "bakchodi"
    ]
}

gaali_count = {}

for gaali in gaali_dict:
    gaali_count[gaali] = 0

for message in normal_messages:

    words = message["message"].lower().split()

    for word in words:

        word = word.strip(string.punctuation)

        for gaali in gaali_dict:

            if word in gaali_dict[gaali]:

                gaali_count[gaali] += 1
                break



print("\n" + "=" * 90)
print(" " * 28 + "🌶️ SPICE ANALYSIS")
print("=" * 90)

print()

print(f"{'Rank':<6}{'Gaali':<15}{'Count':<10}Frequency")
print("-" * 90)

rank = 1

total_gaalis = 0

for gaali in sorted(gaali_count, key=gaali_count.get, reverse=True):

    count = gaali_count[gaali]

    total_gaalis += count

    if count > 0:

        bar = "█" * min(30, count // 5 + 1)

        print(f"{rank:<6}{gaali:<15}{count:<10}{bar}")
        print()

        rank += 1

print("-" * 90)

print(f"🌶️ Total Gaalis : {total_gaalis}")

if total_gaalis < 50:
    spice = "🟢 Mild"

elif total_gaalis < 200:
    spice = "🟡 Medium"

elif total_gaalis < 500:
    spice = "🟠 Spicy"

else:
    spice = "🔴 Nuclear"

print(f"🔥 Spice Level  : {spice}")

print("\n" + "=" * 90)
```

## 😂 Emoji Analysis

```python
emoji_count = {}

emoji_ranges = [
    (0x1F600, 0x1F64F),  # Emoticons
    (0x1F300, 0x1F5FF),  # Symbols & Pictographs
    (0x1F680, 0x1F6FF),  # Transport & Map
    (0x1F900, 0x1F9FF),  # Supplemental Symbols
    (0x2600, 0x26FF),    # Misc Symbols
    (0x2700, 0x27BF)     # Dingbats
]

for message in normal_messages:

    for character in message["message"]:

        unicode_value = ord(character)

        for start, end in emoji_ranges:

            if start <= unicode_value <= end:

                if character in emoji_count:
                    emoji_count[character] += 1
                else:
                    emoji_count[character] = 1

                break





print("\n" + "=" * 90)
print(" " * 32 + "😂 EMOJI ANALYSIS")
print("=" * 90)

print()

if len(emoji_count) == 0:

    print("No emojis found.")

else:

    print(f"{'Rank':<6}{'Emoji':<10}{'Count':<10}Frequency")
    print("-" * 90)

    sorted_emojis = sorted(
        emoji_count.items(),
        key=lambda item: item[1],
        reverse=True
    )

    max_count = sorted_emojis[0][1]

    rank = 1

    for emoji_symbol, count in sorted_emojis[:10]:

        bar = "█" * int((count / max_count) * 25)

        print(f"{rank:<6}{emoji_symbol:<10}{count:<10}{bar}")
        print()

        rank += 1

    print("-" * 90)

    print(f"😂 Total Unique Emojis : {len(emoji_count)}")
    print(f"🏆 Favourite Emoji     : {sorted_emojis[0][0]} ({sorted_emojis[0][1]} times)")

print("\n" + "=" * 90)
```

## 🧠 Chat Vocabulary Analysis

```python
stop_words = [
    "che","to","ne","ma","na","ni","pan","aa","ke","tu","su",
    "ha","haa","hoi","chu","tane","mane","tara","mara","ane",
    "aj","j","kai","koi","je","thi","par","pr","em","cho",
    "hu","a","e","etle","apde","have","tya","ahi","kem"
]

word_count = {}

for message in normal_messages:

    words = message["message"].lower().split()

    for word in words:

        word = word.strip(string.punctuation)

        if word == "":
            continue

        if word.isdigit():
            continue

        if word in stop_words:
            continue

        if len(word) < 3:
            continue

        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1




print("\n" + "=" * 90)
print(" " * 27 + "🧠 CHAT VOCABULARY")
print("=" * 90)

print()

sorted_words = sorted(
    word_count.items(),
    key=lambda item: item[1],
    reverse=True
)

print(f"{'Rank':<6}{'Word':<25}{'Count':<10}Frequency")
print("-" * 90)

max_count = sorted_words[0][1]

rank = 1

for word, count in sorted_words[:20]:

    bar = "█" * int((count / max_count) * 25)

    print(f"{rank:<6}{word:<25}{count:<10}{bar}")
    print()

    rank += 1

print("-" * 90)

print(f"📚 Unique Words : {len(word_count)}")

print("\n" + "=" * 90)
```

## 🧬 Member DNA Analysis

```python
member_stats = {}

for person in participants:

    member_stats[person] = {
        "messages": 0,
        "words": 0,
        "characters": 0,
        "emojis": 0,
        "gaalis": 0,
        "media": 0
    }

for message in messages:

    sender = message["sender"]
    text = message["message"]

    member_stats[sender]["messages"] += 1

    words = text.split()

    member_stats[sender]["words"] += len(words)

    member_stats[sender]["characters"] += len(text)

    for character in text:

        unicode_value = ord(character)

        if (
            0x1F300 <= unicode_value <= 0x1FAFF or
            0x2600 <= unicode_value <= 0x27BF
        ):
            member_stats[sender]["emojis"] += 1

    if "<Media omitted>" in text:
        member_stats[sender]["media"] += 1

    lower_text = text.lower()

    for gaali in gaali_dict:

        for variation in gaali_dict[gaali]:

            if variation in lower_text:
                member_stats[sender]["gaalis"] += 1
                break





print("\n" + "=" * 90)
print(" " * 30 + "🧬 MEMBER DNA")
print("=" * 90)

print()

for person in sorted(message_count, key=message_count.get, reverse=True):

    stats = member_stats[person]

    avg_words = stats["words"] / stats["messages"]

    if avg_words > 15:
        title = "📖 Paragraph Writer"

    elif stats["gaalis"] > 100:
        title = "🌶️ Spice Master"

    elif stats["emojis"] > 300:
        title = "😂 Emoji Lover"

    elif stats["media"] > 100:
        title = "📸 Media King"

    elif stats["messages"] > 1000:
        title = "👑 Conversation King"

    else:
        title = "💬 Active Member"

    print(f"👤 {person}")
    print(f"   Messages      : {stats['messages']}")
    print(f"   Average Words : {avg_words:.2f}")
    print(f"   Emojis        : {stats['emojis']}")
    print(f"   Media         : {stats['media']}")
    print(f"   Gaalis        : {stats['gaalis']}")
    print(f"   DNA           : {title}")
    print()

print("=" * 90)
```

## 🏆 ChatDNA Awards

```python
message_count = {}
word_count_member = {}
emoji_count_member = {}
media_count_member = {}
gaali_count_member = {}
character_count = {}
night_messages = {}
celebration_count = {}

celebration_words = [
    "happy", "birthday", "mubarak", "congratulations",
    "congrats", "anniversary", "eid", "ramadan",
    "milad", "nikah", "🎂", "🎉", "🥳", "🎊"
]

for person in participants:

    message_count[person] = 0
    word_count_member[person] = 0
    emoji_count_member[person] = 0
    media_count_member[person] = 0
    gaali_count_member[person] = 0
    character_count[person] = 0
    night_messages[person] = 0
    celebration_count[person] = 0

for message in messages:

    sender = message["sender"]
    text = message["message"]

    message_count[sender] += 1

    words = text.split()

    word_count_member[sender] += len(words)

    character_count[sender] += len(text)

    if "<Media omitted>" in text:
        media_count_member[sender] += 1

    for character in text:

        unicode_value = ord(character)

        if (
            0x1F300 <= unicode_value <= 0x1FAFF or
            0x2600 <= unicode_value <= 0x27BF
        ):
            emoji_count_member[sender] += 1

    lower_text = text.lower()

    for gaali in gaali_dict:

        for variation in gaali_dict[gaali]:

            if variation in lower_text:
                gaali_count_member[sender] += 1
                break

    for word in celebration_words:

        if word.lower() in lower_text:
            celebration_count[sender] += 1

    time = message["timestamp"].split(",")[1].strip()

    dt = datetime.strptime(time, "%I:%M %p")

    if dt.hour >= 22 or dt.hour < 5:
        night_messages[sender] += 1

average_words = {}

for person in participants:

    average_words[person] = (
        word_count_member[person] /
        message_count[person]
    )




print("\n" + "=" * 90)
print(" " * 31 + "🏆 CHATDNA AWARDS")
print("=" * 90)

print()

winner = max(message_count, key=message_count.get)
print("👑 Conversation King")
print(f"   {winner}")
print(f"   {message_count[winner]} Messages\n")

winner = max(message_count, key=message_count.get)
print("🔥 Most Active Member")
print(f"   {winner}")
print(f"   {message_count[winner]} Messages\n")

winner = max(average_words, key=average_words.get)
print("📖 Paragraph Writer")
print(f"   {winner}")
print(f"   {average_words[winner]:.1f} Words / Message\n")

winner = max(word_count_member, key=word_count_member.get)
print("💬 Word Machine")
print(f"   {winner}")
print(f"   {word_count_member[winner]} Words\n")

winner = max(emoji_count_member, key=emoji_count_member.get)
print("😂 Emoji Lover")
print(f"   {winner}")
print(f"   {emoji_count_member[winner]} Emojis\n")

winner = max(media_count_member, key=media_count_member.get)
print("📸 Media King")
print(f"   {winner}")
print(f"   {media_count_member[winner]} Media Messages\n")

winner = max(gaali_count_member, key=gaali_count_member.get)
print("🌶️ Spice Master")
print(f"   {winner}")
print(f"   {gaali_count_member[winner]} Gaalis\n")

winner = max(night_messages, key=night_messages.get)
print("🌙 Night Owl")
print(f"   {winner}")
print(f"   {night_messages[winner]} Night Messages\n")

winner = max(celebration_count, key=celebration_count.get)
print("🎉 Celebration King")
print(f"   {winner}")
print(f"   {celebration_count[winner]} Celebration Messages\n")

print("=" * 90)
```

## 💬 Conversation Insights

```python
conversation_count = 1
longest_conversation = 1
current_conversation = 1

longest_gap = 0

for i in range(1, len(messages)):

    current = datetime.strptime(messages[i]["timestamp"], "%d/%m/%y, %I:%M %p")
    previous = datetime.strptime(messages[i-1]["timestamp"], "%d/%m/%y, %I:%M %p")

    difference = (current - previous).total_seconds() / 60

    if difference <= 30:
        current_conversation += 1
    else:
        conversation_count += 1

        if current_conversation > longest_conversation:
            longest_conversation = current_conversation

        current_conversation = 1

    if difference > longest_gap:
        longest_gap = difference

if current_conversation > longest_conversation:
    longest_conversation = current_conversation





print("\n" + "=" * 90)
print(" " * 28 + "💬 CONVERSATION INSIGHTS")
print("=" * 90)

print()

print(f"💬 Total Conversations      : {conversation_count}")
print(f"🔥 Longest Conversation     : {longest_conversation} Messages")
print(f"😴 Longest Silent Gap       : {longest_gap / 60:.1f} Hours")

average = len(messages) / conversation_count

print(f"📊 Average Conversation Size: {average:.1f} Messages")

print()

if average > 50:
    print("🧬 Group Type : Highly Interactive")

elif average > 20:
    print("🧬 Group Type : Active")

else:
    print("🧬 Group Type : Casual")

print("\n" + "=" * 90)
```

## 🧬 Chat Personality

```python
total_messages = len(messages)

total_emojis = sum(emoji_count.values())

total_gaalis = sum(gaali_count.values())

night_total = sum(night_messages.values())

celebration_total = sum(celebration_count.values())

emoji_percentage = (total_emojis / total_messages) * 100

gaali_percentage = (total_gaalis / total_messages) * 100

night_percentage = (night_total / total_messages) * 100

celebration_percentage = (celebration_total / total_messages) * 100




print("\n" + "=" * 90)
print(" " * 31 + "🧬 CHAT PERSONALITY")
print("=" * 90)

print()

if emoji_percentage > 40:
    humor = "🤣 Legendary"

elif emoji_percentage > 20:
    humor = "😂 High"

elif emoji_percentage > 10:
    humor = "🙂 Moderate"

else:
    humor = "😐 Low"

if gaali_percentage > 10:
    spice = "🔴 Nuclear"

elif gaali_percentage > 5:
    spice = "🟠 Spicy"

elif gaali_percentage > 2:
    spice = "🟡 Medium"

else:
    spice = "🟢 Mild"

if night_percentage > 30:
    activity = "🌙 Night Owls"

elif night_percentage > 15:
    activity = "🌆 Evening Lovers"

else:
    activity = "🌞 Daytime Chatters"

if celebration_percentage > 5:
    celebration = "🎉 Very High"

elif celebration_percentage > 2:
    celebration = "🥳 High"

else:
    celebration = "🙂 Normal"

friendship_score = (
    50
    + min(15, emoji_percentage)
    + min(15, celebration_percentage)
    + min(20, average)
)

if friendship_score > 100:
    friendship_score = 100

print(f"😂 Humor Level        : {humor}")
print(f"🌶️ Spice Level        : {spice}")
print(f"🌙 Activity Style     : {activity}")
print(f"🎉 Celebration Level  : {celebration}")
print(f"❤️ Friendship Score   : {friendship_score:.0f}/100")

print()

if friendship_score >= 90:
    dna = "🔥 Loud, Active & Chaotic"

elif friendship_score >= 75:
    dna = "😄 Friendly & Interactive"

elif friendship_score >= 60:
    dna = "🙂 Casual Friends"

else:
    dna = "😶 Quiet Group"

print(f"🧬 Overall DNA        : {dna}")

print("\n" + "=" * 90)
```
